# Weather file error

```python
import pandas as pd

error = pd.read_csv("error_weatherfiles.csv", header=[6], index_col=[0], parse_dates=True) #Hourly events on analytics
error.head() 
```

## Historical weather file - error

```python
historicalp = error[['ERROR historical']] # fetching only inputs
historicalp.head()
```

```python
historicalp['Time'] = historicalp.index.map(lambda t: t.time())
historicalp['Date'] = historicalp.index.map(lambda t: t.date())
historicalp.head()
```

```python
dailyhistoricalp = pd.pivot_table(historicalp, values='ERROR historical', index='Date', columns='Time', aggfunc='mean')
dailyhistoricalp.head()
#dailyhistoricalp.T.to_excel("dailyhistoricalp_T.xlsx") 
#dailyhistoricalp.T.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
```

```python
#sns.set_context(context='talk', font_scale=.8)
#styles2 = ['Purple','Orange','LightCoral','LightGreen','r','Plum','Gold','b']


dailyhistoricalp.T.plot(kind='bar', legend=False, figsize=(15,4), alpha=.9, xticks=np.arange(0, 86400, 10800)) # put c='k' for black profiles, style=styles2 for colorful ones


plt.title("Error - Historical - Daily Profiles")
plt.xlabel("Time of the day")
plt.ylabel("Errors (W/m2)")
plt.savefig("test_errors_historical_dailyprofile.png")
```

## Ineichen weather file - error

```python
ineichen = error[['ERROR Inehchen 4']] # fetching only inputs
ineichen.head()
```

```python
ineichen['Time'] = ineichen.index.map(lambda t: t.time())
ineichen['Date'] = ineichen.index.map(lambda t: t.date())
```

```python
dailyineichen = pd.pivot_table(ineichen, values='ERROR Inehchen 4', index='Date', columns='Time', aggfunc='mean')
#dailyineichen.head()
#dailyhistoricalp.T.to_excel("dailyhistoricalp_T.xlsx") 
dailyineichen.T.head()
```

```python
dailyineichen.T.plot(kind='bar', legend=False, figsize=(15,5), alpha=0.7, linewidth=3, xticks=np.arange(0, 86400, 10800)) # put c='k' for black profiles, style=styles2 for colorful ones


plt.title("Error - Ineichen - Daily Profiles")
plt.xlabel("Time of the day")
plt.ylabel("Errors (W/m2)")
plt.savefig("test_errors_ineichen_dailyprofile.png")
```
