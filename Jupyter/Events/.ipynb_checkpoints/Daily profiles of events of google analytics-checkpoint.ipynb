# Visualizing daily profiles - Analytics events

```python
import pandas as pd

events = pd.read_csv("AnalyticsEvents20190311-20190320.csv", header=[6], index_col=[0], parse_dates=True) #Hourly events on analytics
events.head()
```

From day 11.03.2019 to.... (fill last date here)
*HourIndex 0 is equivalent of hour 00:00 to 01:00. 

```python
events['Time'] = events.index.map(lambda t: t.time())
events['Date'] = events.index.map(lambda t: t.date())
events.head()
```

```python
dailyblocks = pd.pivot_table(events, values='Total Events', index='Date2', columns='Time', aggfunc='mean')
dailyblocks.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns

sns.set_context(context='talk', font_scale=.8)
styles2 = ['Purple','Orange','LightCoral','LightGreen','r','Plum','Gold','b']

dailyblocks.T.plot(legend=False, figsize=(15,5), style=styles2, alpha=0.4, xticks=np.arange(0, 86400, 10800)) # put <c='k'> for black profiles 

plt.title("Events - Daily Profiles")
plt.xlabel("Time of the day")
plt.ylabel("Events")
plt.savefig("test_events_dailyprofile.png", dpi=300, papertype=None, format=None,
        transparent=True, bbox_inches=None, pad_inches=0.1,
        frameon=None, metadata=None)
```
